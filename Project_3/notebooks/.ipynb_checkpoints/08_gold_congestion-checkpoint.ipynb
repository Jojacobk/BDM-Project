```python
import os
os.environ['JAVA_HOME'] = '/usr/lib/jvm/java-17-openjdk-amd64'
os.environ['PYSPARK_SUBMIT_ARGS'] = '--packages org.apache.spark:spark-sql-kafka-0-10_2.13:4.1.0,org.apache.iceberg:iceberg-spark-runtime-4.0_2.13:1.10.0,org.apache.iceberg:iceberg-aws-bundle:1.10.0 pyspark-shell'
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql import Window
```

```python
spark = (
    SparkSession.builder
    .appName("gold_congestion")
    .config("spark.sql.extensions", "org.apache.iceberg.spark.extensions.IcebergSparkSessionExtensions")
    .config("spark.sql.catalog.lakehouse", "org.apache.iceberg.spark.SparkCatalog")
    .config("spark.sql.catalog.lakehouse.type", "rest")
    .config("spark.sql.catalog.lakehouse.uri", "http://iceberg-rest:8181")
    .config("spark.sql.catalog.lakehouse.io-impl", "org.apache.iceberg.aws.s3.S3FileIO")
    .config("spark.sql.catalog.lakehouse.s3.endpoint", "http://minio:9000")
    .config("spark.sql.catalog.lakehouse.s3.path-style-access", "true")
    .config("spark.sql.catalog.lakehouse.warehouse", "s3://warehouse/")
    .config("spark.sql.defaultCatalog", "lakehouse")
    .config("spark.sql.shuffle.partitions", "4")
    .getOrCreate()
)
spark.sparkContext.setLogLevel("WARN")
```

```python
SILVER_TABLE = "lakehouse.taxi.silver"
IMPACT_TABLE = "lakehouse.taxi.gold_congestion_impact"
ZONES_TABLE  = "lakehouse.taxi.gold_congestion_zones"
```

```python
def ensure_tables(spark):
    spark.sql("CREATE NAMESPACE IF NOT EXISTS lakehouse.taxi")
    spark.sql("""
        CREATE TABLE IF NOT EXISTS lakehouse.taxi.gold_congestion_impact (
            PULocationID             INT,
            pickup_zone              STRING,
            hour_of_day              INT,
            avg_speed_mph            DOUBLE,
            avg_congestion_surcharge DOUBLE,
            trips_with_surcharge     LONG,
            trips_without_surcharge  LONG,
            total_congestion_revenue DOUBLE,
            avg_fare_per_mile        DOUBLE,
            trip_date                DATE
        )
        USING iceberg
        PARTITIONED BY (trip_date)
        TBLPROPERTIES ('write.target-file-size-bytes' = '67108864')
    """)
    spark.sql("""
        CREATE TABLE IF NOT EXISTS lakehouse.taxi.gold_congestion_zones (
            most_congested_zones STRING,
            top_revenue_zones    STRING,
            speed_profile_top3   STRING,
            trip_date            DATE
        )
        USING iceberg
        PARTITIONED BY (trip_date)
        TBLPROPERTIES ('write.target-file-size-bytes' = '67108864')
    """)
    print("[gold_congestion] Tables ready.")
```

```python
def run_impact(spark):
    silver_df = spark.read.format("iceberg").load(SILVER_TABLE)

    impact_df = (
        silver_df
        .withColumn("hour_of_day", F.hour(F.col("tpep_pickup_datetime")))
        .groupBy("PULocationID", "pickup_zone", "trip_date", "hour_of_day")
        .agg(
            F.avg(
                F.when(
                    F.col("trip_duration_minutes") > 0,
                    F.col("trip_distance") / (F.col("trip_duration_minutes") / 60.0),
                )
            ).alias("avg_speed_mph"),
            F.avg("congestion_surcharge").alias("avg_congestion_surcharge"),
            F.count(F.when(F.col("congestion_surcharge") > 0, F.lit(1))).cast("long").alias("trips_with_surcharge"),
            F.count(F.when(F.col("congestion_surcharge") == 0, F.lit(1))).cast("long").alias("trips_without_surcharge"),
            F.sum("congestion_surcharge").alias("total_congestion_revenue"),
            F.avg(
                F.when(F.col("trip_distance") > 0, F.col("fare_amount") / F.col("trip_distance"))
            ).alias("avg_fare_per_mile"),
        )
    )

    (
        impact_df
        .writeTo(IMPACT_TABLE)
        .using("iceberg")
        .overwritePartitions()
    )

    print("[gold_congestion] Impact table written.")
```

```python
def run_zones(spark):
    impact_df = spark.read.format("iceberg").load(IMPACT_TABLE)

    daily_zone_speed = (
        impact_df
        .groupBy("trip_date", "pickup_zone")
        .agg(F.avg("avg_speed_mph").alias("zone_avg_speed"))
    )
    speed_window = Window.partitionBy("trip_date").orderBy(F.asc("zone_avg_speed"))
    daily_zone_speed = daily_zone_speed.withColumn("speed_rank", F.row_number().over(speed_window))

    most_congested = (
        daily_zone_speed
        .filter(F.col("speed_rank") <= 5)
        .groupBy("trip_date")
        .agg(F.to_json(F.collect_list("pickup_zone")).alias("most_congested_zones"))
    )

    daily_zone_revenue = (
        impact_df
        .groupBy("trip_date", "pickup_zone")
        .agg(F.sum("total_congestion_revenue").alias("zone_total_revenue"))
    )
    revenue_window = Window.partitionBy("trip_date").orderBy(F.desc("zone_total_revenue"))
    daily_zone_revenue = daily_zone_revenue.withColumn("revenue_rank", F.row_number().over(revenue_window))

    top_revenue = (
        daily_zone_revenue
        .filter(F.col("revenue_rank") <= 5)
        .groupBy("trip_date")
        .agg(F.to_json(F.collect_list("pickup_zone")).alias("top_revenue_zones"))
    )

    top3_zones = (
        daily_zone_speed
        .filter(F.col("speed_rank") <= 3)
        .select("trip_date", "pickup_zone")
    )

    top3_hourly = (
        impact_df
        .join(top3_zones, ["trip_date", "pickup_zone"])
        .select("trip_date", "pickup_zone", "hour_of_day", "avg_speed_mph")
    )

    speed_profile_top3 = (
        top3_hourly
        .groupBy("trip_date", "pickup_zone")
        .agg(
            F.to_json(
                F.collect_list(F.struct("hour_of_day", "avg_speed_mph"))
            ).alias("hourly_profile")
        )
        .groupBy("trip_date")
        .agg(
            F.to_json(
                F.collect_list(F.struct("pickup_zone", "hourly_profile"))
            ).alias("speed_profile_top3")
        )
    )

    zones_df = (
        most_congested
        .join(top_revenue, "trip_date", "outer")
        .join(speed_profile_top3, "trip_date", "outer")
    )

    (
        zones_df
        .writeTo(ZONES_TABLE)
        .using("iceberg")
        .overwritePartitions()
    )

    print("[gold_congestion] Zones table written.")
```

```python
def run_queries(spark):
    spark.sql("""
        SELECT pickup_zone, AVG(avg_speed_mph) AS rush_hour_speed
        FROM lakehouse.taxi.gold_congestion_impact
        WHERE hour_of_day = 8
        GROUP BY pickup_zone
        ORDER BY rush_hour_speed ASC
        LIMIT 1
    """).show(truncate=False)

    spark.sql("""
        SELECT trip_date, SUM(total_congestion_revenue) AS daily_revenue
        FROM lakehouse.taxi.gold_congestion_impact
        GROUP BY trip_date
        ORDER BY trip_date
    """).show(truncate=False)
```

```python
ensure_tables(spark)
run_impact(spark)
run_zones(spark)
run_queries(spark)
```
